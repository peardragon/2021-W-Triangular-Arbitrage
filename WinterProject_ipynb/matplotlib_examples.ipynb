```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,100, 100).tolist()
y = np.sin(x).tolist()
```

```python
x, y
```

```python
fig, axs = plt.subplots(2,1)
y = np.array(y)
x = np.array(x)
axs[0].plot( y)
axs[0].set_title("Free Balance")
axs[1].plot(x)
axs[1].set_title("Total Balance")

fig.tight_layout()
plt.show()
```
