```python
%%time
import urllib.request
import json
import datetime


import pandas as pd


def timedelta_datetime_min(date_time,delta_):
    """
    :return: add 1 minute datetime
    """
    import datetime
    date_time = datetime.datetime.strptime(date_time, "%Y-%m-%d %H:%M:%S")
    date_time += datetime.timedelta(minutes=delta_)
    return date_time


def timedelta_datetime_day(date_time,delta_):
    """
    :return: add 1 minute datetime
    """
    import datetime
    date_time = datetime.datetime.strptime(date_time, "%Y-%m-%d %H:%M:%S")
    date_time += datetime.timedelta(days=delta_)
    return date_time

def convert_to_unixtime(date_time):
    """Convert datetime to unixtime"""
    unixtime = datetime.datetime.strptime(date_time, "%Y-%m-%d %H:%M:%S").timestamp()
    return unixtime


```

```python
class feature_selection:

  def __init__(self,data_input):

    self.coinbase_data = data_input

    self.data_added_all_ta = ta.add_all_ta_features(self.coinbase_data, 'BTC:open', 'BTC:high', 'BTC:low', 'BTC:close', 'BTC:volume')
    self.volume_ta = self.data_added_all_ta.loc[:, [name.startswith("volume") for name in self.data_added_all_ta.columns]]
    self.volatility_ta = self.data_added_all_ta.loc[:, [name.startswith("volatility") for name in self.data_added_all_ta.columns]]
    self.trend_ta = self.data_added_all_ta.loc[:, [name.startswith("trend") for name in self.data_added_all_ta.columns]]
    self.momentum_ta = self.data_added_all_ta.loc[:, [name.startswith("momentum") for name in self.data_added_all_ta.columns]]

    self.trend_corr_matrix = self.trend_ta.corr()
    self.volatility_corr_matrix = self.volatility_ta.corr()
    self.momentum_corr_matrix = self.momentum_ta.corr()
    self.volume_corr_matrix = self.volume_ta.corr()

  def feature_selection_corr_matrix(self, corr, data):
    import numpy as np
    columns = np.full((corr.shape[0],), True, dtype=bool)
    for i in range(corr.shape[0]):
        for j in range(i+1, corr.shape[0]):
            if corr.iloc[i,j] >= 0.8:
                if columns[j]:
                    columns[j] = False
    self.selected_columns = data.columns[columns]
    return self.selected_columns

  def get_selected_index(self):
    self.col_selected = self.feature_selection_corr_matrix(self.trend_corr_matrix, self.trend_ta)
    self.col_selected = self.col_selected.append(self.feature_selection_corr_matrix(self.volatility_corr_matrix, self.volatility_ta))
    self.col_selected = self.col_selected.append(self.feature_selection_corr_matrix(self.momentum_corr_matrix, self.momentum_ta))
    self.col_selected = self.col_selected.append(self.feature_selection_corr_matrix(self.volume_corr_matrix, self.volume_ta))

    return self.col_selected
  def return_selected_data(self):
    columns = self.get_selected_index()
    self.data_feature_selected = self.data_added_all_ta[columns]
    return self.data_feature_selected
```
